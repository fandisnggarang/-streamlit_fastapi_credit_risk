```python
# Import utils dari src dan RandomOverSampler dari imblearn.over_sampling
from src import utils
from imblearn.over_sampling import RandomOverSampler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Muat data train set yang telah dilakukan preprocessing
X_train_prep = utils.deserialize_data('data/processed/X_train_prep.pkl')
y_train_prep = utils.deserialize_data('data/processed/y_train_prep.pkl')
```

```python
# Buat instance RandomOverSampler
ros = RandomOverSampler()
```

```python
# Ubah y_train_prep ke DataFrame
y_train_prep_df = pd.DataFrame(y_train_prep, columns=["loan_status"])

# Buat grafik countplot
plt.figure(figsize=(10, 6))
sns.countplot(data=y_train_prep_df, x="loan_status")
plt.title("Distribusi Loan Status Sebelum Oversampling")
plt.xlabel("Loan Status")
plt.ylabel("Count")
plt.show()

```

```python
# Lakukan oversampling
X_train_ros, y_train_ros = ros.fit_resample(X_train_prep, y_train_prep)
```

```python
# Ubah y_train_ros menjadi DataFrame
y_train_ros_df = pd.DataFrame(y_train_ros, columns=['loan_status'])

# Buat grafik kondisi label setelah oversampling
plt.figure(figsize=(8, 6))
sns.countplot(data=y_train_ros_df, x='loan_status', hue='loan_status')
plt.title('Distribusi Label Setelah Oversampling')
plt.show()
```

```python

# Serialisasi X_train_ros
utils.serialize_data(X_train_ros, 'data/processed/X_train_ros.pkl')


```
